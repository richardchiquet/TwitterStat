```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
import numpy as np
```

```python
data = pd.read_csv("dataV1.csv",sep="|")
```

```python
dates=[]
lenght=[]
means=[]
Max_on_one_word=[]
```

```python
for i in data.index:
    Y = int(data["date_time"][i].split()[0][6:-1])
    M = int(data["date_time"][i].split()[0][:2])
    D = int(data["date_time"][i].split()[0][3:5])
    h = int(data["date_time"][i].split()[1][0:2])
    m = int(data["date_time"][i].split()[1][3:5])
    dates.append(datetime(Y,M,D,h,m))
    lenght.append(int(data["lenght"][i]))
    means.append(float(data["means"][i]))
    Max_on_one_word.append(int(data["Max_on_one_word"][i]))
```

```python

plt.style.use('seaborn')

plt.plot_date(dates,lenght, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('leght')
plt.xlabel('Date')
plt.ylabel('leght')

plt.tight_layout()

plt.show()
plt.close()
```

```python
plt.style.use('seaborn')

plt.plot_date(dates,means, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('means')
plt.xlabel('Date')
plt.ylabel('means')

plt.tight_layout()

plt.show()
plt.close()
```

```python
plt.style.use('seaborn')

plt.plot_date(dates,Max_on_one_word, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Max_on_one_word')
plt.xlabel('Date')
plt.ylabel('Max_on_one_word')

plt.tight_layout()

plt.show()
plt.close()
```
